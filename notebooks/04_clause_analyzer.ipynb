# 04 – Clause Analyzer

```python
# config: store, retrieval, clause analyzer agent, and query
import sys
from pathlib import Path

ROOT = Path("../").resolve()
sys.path.insert(0, str(ROOT))

from rag.config import get_settings
from rag.retrieval.store import IndexStore
from rag.retrieval.hybrid_search import HybridRetrieval
from rag.agents.clause_analyzer import ClauseAnalyzerAgent

settings = get_settings()
store = IndexStore(persist_directory=ROOT / settings.persist_directory)
retrieval = HybridRetrieval(store=store, top_k=5)
clause_agent = ClauseAnalyzerAgent()

QUERY = "What are the confidentiality obligations and do they survive termination?"
```

```python
# retrieve chunks, run clause analyzer, print obligations, exceptions, caps, survival, remedies
retrieved = retrieval.search(QUERY)
analysis = clause_agent.analyze(retrieved)
print("Obligations:", analysis.obligations)
print("Exceptions:", analysis.exceptions)
print("Caps:", analysis.caps)
print("Survival:", analysis.survival_clauses)
print("Remedies:", analysis.remedies)
if analysis.raw_summary:
    print("Summary:", analysis.raw_summary)
```
