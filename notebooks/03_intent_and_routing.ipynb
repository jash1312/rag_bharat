# 03 – Intent & Routing

```python
# config: intent agent and sample queries (incl. legal-advice and vague)
import sys
from pathlib import Path

ROOT = Path("../").resolve()
sys.path.insert(0, str(ROOT))

from rag.agents.intent import IntentAgent

intent_agent = IntentAgent()
QUERIES = [
    "What is the uptime commitment?",
    "Should I sign this NDA?",
    "Tell me about liability.",
]
```

```python
# run intent classification for each query; print type, focus, legal-advice, clarification
for q in QUERIES:
    result = intent_agent.classify(q)
    print(f"Query: {q}")
    print(f"  intent_type={result.intent_type}, document_focus={result.document_focus}")
    print(f"  is_legal_advice={result.is_legal_advice}, requires_clarification={result.requires_clarification}")
    print()
```
