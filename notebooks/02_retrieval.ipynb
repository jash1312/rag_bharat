# 02 – Retrieval

```python
# config: store, hybrid retrieval (BM25 + vector), query and top_k
import sys
from pathlib import Path

ROOT = Path("../").resolve()
sys.path.insert(0, str(ROOT))

from rag.config import get_settings
from rag.retrieval.store import IndexStore
from rag.retrieval.hybrid_search import HybridRetrieval

settings = get_settings()
store = IndexStore(persist_directory=ROOT / settings.persist_directory)
retrieval = HybridRetrieval(
    store=store,
    top_k=settings.top_k,
    bm25_weight=settings.bm25_weight,
    vector_weight=settings.vector_weight,
)

QUERY = "What is the notice period for terminating the NDA?"
TOP_K = 5
```

```python
# run hybrid search and print ranked results with scores
results = retrieval.search(QUERY, top_k=TOP_K)
for i, r in enumerate(results):
    print(f"--- {i+1} [{r.chunk.document} | {r.chunk.title}] score={r.score:.3f}")
    print(r.chunk.clause_text[:400], "...")
    print()
```
