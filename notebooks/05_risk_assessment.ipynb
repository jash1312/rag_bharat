# 05 – Risk Assessment

```python
# config: store, retrieval, risk agent, and query
import sys
from pathlib import Path

ROOT = Path("../").resolve()
sys.path.insert(0, str(ROOT))

from rag.config import get_settings
from rag.retrieval.store import IndexStore
from rag.retrieval.hybrid_search import HybridRetrieval
from rag.agents.risk import RiskAssessmentAgent

settings = get_settings()
store = IndexStore(persist_directory=ROOT / settings.persist_directory)
retrieval = HybridRetrieval(store=store, top_k=5)
risk_agent = RiskAssessmentAgent()

QUERY = "Is liability capped in the NDA?"
```

```python
# retrieve chunks, assess risk, print level, reason, financial exposure, patterns
retrieved = retrieval.search(QUERY)
risk = risk_agent.assess(retrieved)
print("Risk level:", risk.risk_level)
print("Reason:", risk.reason)
print("Financial exposure:", risk.financial_exposure)
print("Patterns:", risk.risk_patterns)
```
