# 01 – Ingestion

```python
# config: project root, paths, and index-to-store flag
import sys
from pathlib import Path

ROOT = Path("../").resolve()
sys.path.insert(0, str(ROOT))

from rag.config import get_settings

settings = get_settings()
INPUT_DIR = settings.input_data_dir
PERSIST_DIR = settings.persist_directory
INDEX_TO_STORE = True

print("Input dir:", INPUT_DIR)
print("Persist dir:", PERSIST_DIR)
```

```python
# run ingestion
from rag.ingestion.pipeline import run_ingestion

chunks = run_ingestion(
    input_dir=ROOT / INPUT_DIR,
    persist_directory=ROOT / PERSIST_DIR,
    index_to_store=INDEX_TO_STORE,
)
print(f"Indexed {len(chunks)} clause chunks.")
```

```python
# show sample chunk from indexed corpus
if chunks:
    c = chunks[0]
    print("Sample chunk:", c.document, c.section, c.title)
    print(c.clause_text[:300], "...")
```
